```python
import numpy as np
import pandas as pd
a=np.random.uniform(low=1,high=100,size=(10,1))
df=pd.DataFrame(a,columns=['voltage'])

df['Target']=50;
x=np.arange(-3*3.14,3*3.14)
y=np.sin(x)
dic={'x':x,'y':y}
df1=pd.DataFrame(dic)
df1.describe()
target=df1.pop('y')
x1=np.arange(4*3.14,5*3.14)
y1=np.sin(x1)
dic1={'x1':x1,'y1':y1}
df2=pd.DataFrame(dic1)
target1=df2.pop('y1')

mf=[[['gaussmf',{'mean':np.mean(np.arange(-10,-3)),'sigma':np.std(np.arange(-10,-3))}]
,['gaussmf',{'mean':np.mean(np.arange(-7,0)),'sigma':np.std(np.arange(-7,0))}]
,['gaussmf',{'mean':np.mean(np.arange(-3,40)),'sigma':np.std(np.arange(-3,40))}]]]

from membership import membershipfunction
mfc = membershipfunction.MemFuncs(mf)

import anfis
anf = anfis.ANFIS(df1,y, mfc)

pred_train=anf.trainHybridJangOffLine(epochs=20)

error=np.mean((pred_train-y)**2)
print(error)

anf.plotErrors()
anf.plotResults()

y_pred=anfis.predict(anf,df2.values)
print(y_pred)

from sklearn.metrics import  mean_squared_error
error=mean_squared_error(y1.reshape(len(y1),1),y_pred)
print(error)
```
